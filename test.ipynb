```python
def get_relevant_info(soup):
    date = soup.find('a', attrs={'class': 'css-146c3p1'}).text
    post_text = soup.find_all('span', attrs={'class': 'css-1jxf684'})[5].text
    
    if soup.find('div', attrs={'data-testid': 'tweetPhoto'}):
        img_html = soup.find_all('img', attrs={'class':'css-9pa8cd'})[1]
        image_url = img_html['src']
    else:
        image_url = None
    return date, post_text, image_url
```

```python
import pandas as pd

data = []
for html in list(scraper.article_htmls.values())[:5]:
    soup = BeautifulSoup(html, 'html.parser')
    date, post_text, image_url = get_relevant_info(soup)
    print(date, post_text, image_url, sep=', ')
    data.append([date, post_text, image_url])
```

# New method

```python
from twitter_scraper import TwitterScraper

scraper = TwitterScraper(
    search_query="space",
    from_account="NASA",
    start_date="2023-11-01",
    end_date="2024-01-31",
    scroll_pause_time=4,
    scroll_attempt_limit=8
)
```

```python
scraper.setup_driver()
```

```python
scraper.load_cookies("cookies.json")
```

```python
articles = scraper.scrape(save_screenshots=False)
```

```python
len(articles)
```

```python
scraper.sorted_articles
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(articles[100])
print(soup.prettify())
```

```python
soup.get_text()
```

```python
soup.find('div', attrs={'data-testid': 'tweetPhoto'})
soup.find_all('img', attrs={'class':'css-9pa8cd'})
```

```python
def get_relevant_info(soup):
    post_text = soup.get_text()
    
    poss_img = soup.find_all('img', attrs={'class':'css-9pa8cd'})
    if len(poss_img) > 1:
        img_html = poss_img[1]
        image_url = img_html['src']
    else:
        image_url = None
    return post_text, image_url
```

```python
get_relevant_info(soup)
```

```python
type(soup)
```

```python
import uuid
import pandas as pd

data = []
for d in scraper.sorted_articles:
    date = d['timestamp']
    article = d['html']

    name = str(uuid.uuid4())
    soup = BeautifulSoup(article)
    post_text, image_url = get_relevant_info(soup)
    data.append((name, date, post_text, image_url))

df = pd.DataFrame(data, columns=['ID', 'Date', 'Post Text', 'Image URL'])
```
